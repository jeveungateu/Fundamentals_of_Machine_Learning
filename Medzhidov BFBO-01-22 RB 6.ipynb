## Задание 1
Сгенерируйте самостоятельно начальные данные для линейной регрессии. Произведите разделение данных на обучающую и тестовую выборку. Обучите модель линейной регрессии на тренировочной выборке и
предскажите значения на тестовой. Вычислите метрики MSE и MAE на тренировочной и тестовой выборках. Сделайте выводы. Верно ли, что ошибка на обучающей выборке всегда меньше, чем на тестовой?

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

noise = np.random.randn(20) * 10

X = np.linspace(-5, 5, 20)
y = 10 * X - 7 + noise
X = X.reshape(-1, 1)
X.shape

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.5)
plt.figure(figsize=(10, 5))
plt.plot(X, 10 * X - 7, label='real')
plt.scatter(X_train, y_train, label='train')
plt.scatter(X_test, y_test, label='test')
plt.legend()
plt.show()

model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

w = model.coef_[0]
b = model.predict(np.zeros((1, 1)))[0]
w, b

plt.figimage
plt.figure(figsize=(10, 5))
plt.plot(X, 10 * X - 7, label='real')
plt.scatter(X_train, y_train, label='train')
plt.scatter(X_test, y_test, label='test')
plt.plot(X, X * w + b, label='predicted')
plt.legend()
plt.show()


y_train_predicted = model.predict(X_train)
y_test_predicted = model.predict(X_test)

print('Train MSE: ', mean_squared_error(y_train, y_train_predicted))
print('Test MAE: ', mean_squared_error(y_test, y_test_predicted))

print('Train MSE: ', mean_absolute_error(y_train, y_train_predicted))
print('Test MAE: ', mean_absolute_error(y_test, y_test_predicted))
```

```python
Сделайте выводы. Верно ли, что ошибка на обучающей выборке всегда меньше, чем на тестовой? 

## Ответ:
Среднеквадратичная ошибка сильнее реагирует на большие отклонения в данных (выбросы значений) по сравнению со средним абсолютным, и поэтому более чувствительна к выбросам.
```

```python
import numpy as np
import matplotlib.pyplot as plt
import requests
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from pprint import pprint

noise = np.random.randn(20) * 10
plt.hist(noise)
def get_info():
    url = 'https://raw.githubusercontent.com/AnnaShestova/salary-years-simple-linear-regression/master/Salary_Data.csv'
    response = requests.get(url).text
    columns = [i.strip() for i in response.split('\n')]
    second_str = columns.pop(0).split(',')
    columns = [i.split(',') for i in columns if i]
    dict_regr = {second_str[0]: [i[0] for i in columns],
                 second_str[1]: [i[1] for i in columns]}
    return dict_regr

plt.scatter(get_info()['YearsExperience'], get_info()['Salary'], color='b', label = 'Данные о заработке')
plt.xlabel("Опыт работы (в годах)")
plt.ylabel("Зарплата")
plt.show()

X = [[float(i)] for i in get_info()['YearsExperience']]
y = [[float(i)] for i in get_info()['Salary']]
pprint(X)
pprint(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5, random_state = 0)

plt.figure(figsize=(10, 5))

plt.scatter(X_train, y_train, label='train')
plt.scatter(X_test, y_test, label='test')
plt.legend()
plt.show()

model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

w = model.coef_[0]
b = model.predict(np.zeros((1, 1)))[0]
print(w,b)

plt.figimage
plt.figure(figsize=(10, 5))
plt.scatter(X_train, y_train, label='train')
plt.scatter(X_test, y_test, label='test')
plt.plot(X, X * w + b, label='predicted')
plt.legend()
plt.show()


y_train_predicted = model.predict(X_train)
y_test_predicted = model.predict(X_test)

print('Train MSE: ', mean_squared_error(y_train, y_train_predicted))
print('Test MAE: ', mean_squared_error(y_test, y_test_predicted))

print('Train MSE: ', mean_absolute_error(y_train, y_train_predicted))
print('Test MAE: ', mean_absolute_error(y_test, y_test_predicted))
```
